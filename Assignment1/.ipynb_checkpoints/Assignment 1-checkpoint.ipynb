### a)

Here we have a function that returns the value of the Lagrange function at a given point x: $L_k(x)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lagrange(x, nodes, k):
    f = 1 # Neutral element for multiplication
    for i in range(0, len(nodes)):
        if i != k:
            f *= (x-nodes[i]) / (nodes[k] - nodes[i])
    return f
```

### b)

Now we perform the interpolation and calculate the sum of the polynomials from our Lagrange function

$p_n(x) = \sum_{k=0}^{n} f(x_k)L_k(x)$

```python
def lagrange_interpolation(x, nodes, function_values):
    if len(nodes) != len(function_values):
        raise("Error: Nodes list must be as long as function_values")
    f = 0 # Neutral element for addition
    for i in range(0, len(nodes)):
        f += function_values[i] * lagrange(x, nodes, i)
    return f
```

### c)

Finally, we'll plot how our function performs on the sine function based on $n \in 1,2,3,4$ and nodes defined as $x_k = \frac{k}{n} \pi, k = 0, ..., n$

```python
# Assignment c

# The function we are interpolating
def sin_function(x):
    return np.sin(x)

ns = [1,2,3,4]

interval = np.linspace(0, np.pi, 100)

plt.plot(interval, sin_function(interval), label="sin(x)")

for n in ns:
    nodes = np.linspace(0, np.pi, n+1)

    sin_vals = sin_function(nodes)

    interpolation = lagrange_interpolation(interval, nodes, sin_vals)

    plt.plot(interval, interpolation, label=f'Lagrange Interpolation (n={n})')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Interpolation of the sine function')
plt.legend()
plt.grid(True)
plt.show()
```
